```python
def load_data(data_dir):
  images = []
  labels = []

  for filename in os.listdir(os.path.join(data_dir, 'Images')):
    if filename.endswith('.png'):
      img_path = os.path.join(data_dir, 'Images', filename)
      label_path = os.path.join(data_dir, 'Labels', filename)
      images.append(img_path)
      labels.append(label_path)

  return images, labels
```

```python
train_image, train_labels = load_data('Dataset_traffic/train_data')
test_image, test_labels = load_data('Dataset_traffic/test_data')
val_image, val_labels = load_data('Dataset_traffic/val_data')
```

```python

def create_dataset(image_path, labels_path, test = False):
  if test:
    dataset = Dataset.from_dict({'image':sorted(image_path)})
    dataset = dataset.cast_column('image', Image())
    return dataset

  dataset = Dataset.from_dict({'image':sorted(image_path),
                               'label': sorted(labels_path)})
  dataset = dataset.cast_column('image', Image())
  dataset = dataset.cast_column('label', Image())
  return dataset
```

```python
train_dataset = create_dataset(train_image, train_labels)
test_dataset = create_dataset(test_image, test_labels, test = True)
val_dataset = create_dataset(val_image, val_labels)
```

```python
dataset = DatasetDict({
     "train": train_dataset,
     "validation": val_dataset
     }
)
```

```python
dataset.push_to_hub("aslez123/street_dataset", private=True)
```
